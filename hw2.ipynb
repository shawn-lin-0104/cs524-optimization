```julia
using JuMP, HiGHS

m = Model(HiGHS.Optimizer)

@variable(m, x >= 3) # process 1
@variable(m, y >= 0) # process 2

@constraint(m, 3x + y >= 10) # vials of A constraints
@constraint(m, x + y >= 5) # vials of B constraints

@objective(m, Min, 4x + y) # total cost

optimize!(m)

println("optimal use time of process 1: ", value(x))
println("optimal use time of process 2: ", value(y))

println("optimal cost: \$", objective_value(m))
```

```julia
using JuMP, HiGHS

m = Model(HiGHS.Optimizer)

I = [:Alkylate, :Catalytic_Cracked, :Straight_Run, :Isopentane]  
J = [:R, :gasA, :gasB]  

PN = Dict(:Alkylate => 107, :Catalytic_Cracked => 93, :Straight_Run => 87, :Isopentane => 108)
RV_P = Dict(:Alkylate => 5, :Catalytic_Cracked => 8, :Straight_Run => 4, :Isopentane => 21)
barrel = Dict(:Alkylate => 3814, :Catalytic_Cracked => 2666, :Straight_Run => 4016, :Isopentane => 1300)

price = Dict(:R => 4.83, :gasA => 6.45, :gasB => 5.91)

@variable(m, x[I, J] >= 0)


@constraint(m, [i in I], sum(x[i, j] for j in J) == barrel[i]) # Barrels produce limits
@constraint(m, sum( (PN[i]-100) * x[i, :gasA] for i in I) >= 0)  # PN Avgas A constraints
@constraint(m, sum( (PN[i]-91) * x[i, :gasB] for i in I) >= 0)  # PN Avgas B constraints
@constraint(m, sum( (RV_P[i]-7) * x[i, :gasA] for i in I) <= 0)  # RV P Avgas A constraints
@constraint(m, sum( (RV_P[i]-7) * x[i, :gasB] for i in I) <= 0)  # RV P Avgas A constraints

@objective(m, Max, sum(price[j] * x[i, j] for i in I, j in J))

optimize!(m)

println("optimal profit: ", objective_value(m))
for i in I, j in J
    println(value(x[i, j]), " barrels of ", i, " contribute to ", j)
end
```

```julia
using JuMP, HiGHS

E = [ :C,  :Cu, :Mn]
J = [ :F,  :U,  :E,  :L,  :O,  :N,  :M]

c = Dict( :F => 200,  :U => 250,  :E => 150,  :L => 220,  :O => 240,  :N => 200,  :M => 165)

u = Dict( :F => 400,  :U => 300,  :E => 600,  :L => 500,  :O => 200,  :N => 300,  :M => 250)

γ = Dict(
     :C => Dict( :F => 2.5,  :U => 3.0,  :E => 1.5,  :L => 0.0,  :O => 0.4,  :N => 0.0,  :M => 0.0),
     :Cu => Dict( :F => 0.0,  :U => 0.3,  :E => 0.5,  :L => 0.9,  :O => 0.9,  :N => 0.4,  :M => 0.6),
     :Mn => Dict( :F => 1.3,  :U => 1.8,  :E => 1.5,  :L => 0.0,  :O => 1.1,  :N => 1.2,  :M => 0.0)
)

α = Dict( :C => 2,  :Cu => 0.4,  :Mn => 1.2)
β = Dict( :C => 3,  :Cu => 0.6,  :Mn => 1.65)

model = Model(HiGHS.Optimizer)

@variable(model, d[J] >= 0)

@constraint(model, d["F"] <= u["F"])

@objective(model, Min, sum(c[j] * d[j] for j in J))

@constraint(model, sum(d[j] for j in J) == 500)

for e in E
    @constraint(model, sum(γ[e][j] * d[j] for j in J) / 500 <= β[e])
    @constraint(model, sum(γ[e][j] * d[j] for j in J) / 500 >= α[e])
end

optimize!(model)

println( :Termination Status: , termination_status(model))
println( :Optimal Cost: \$, objective_value(model))
```

```julia
using JuMP, HiGHS

E = [:C, :Cu, :Mn]
J = [:F, :U, :E, :L, :O, :N, :M] 

c = Dict(:F => 200, :U => 250, :E => 150, :L => 220, :O => 240, :N => 200, :M => 165)

u = Dict(:F => 400, :U => 300, :E => 600, :L => 500, :O => 200, :N => 300, :M => 250)

γ = Dict(
    :C  => Dict(:F => 2.5, :U => 3.0, :E => 1.5, :L => 0.0, :O => 0.4, :N => 0.0, :M => 0.0),
    :Cu => Dict(:F => 0.0, :U => 0.3, :E => 0.5, :L => 0.9, :O => 0.9, :N => 0.4, :M => 0.6),
    :Mn => Dict(:F => 1.3, :U => 1.8, :E => 1.5, :L => 0.0, :O => 1.1, :N => 1.2, :M => 0.0)
)

α = Dict(:C => 2, :Cu => 0.4, :Mn => 1.2)
β = Dict(:C => 3, :Cu => 0.6, :Mn => 1.65)

model = Model(HiGHS.Optimizer)

@variable(model, 0 <= x[j in J] <= u[j])

@objective(model, Min, sum(c[j] * x[j] for j in J))

@constraint(model, sum(x[j] for j in J) == 500)

for e in E
    @constraint(model, α[e] <= sum(γ[e][j] * x[j] for j in J) / 500)
    @constraint(model, sum(γ[e][j] * x[j] for j in J) / 500 <= β[e])
end

optimize!(model)

println("Termination Status: ", termination_status(model))
println("optimal objective function value: ", objective_value(model))
```
