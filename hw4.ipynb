```julia
using JuMP, HiGHS

m = Model(HiGHS.Optimizer)

@variable(m, x1 >= 0) 
@variable(m, x2 >= 0) 

@constraint(m, 4x1 + x2 <= 7) # vials of A constraints
@constraint(m, 5x1 + 2x2 <= 15) # vials of B constraints

@objective(m, Max, 3x1 + x2) # total cost

optimize!(m)

println("optimal objective value: ", objective_value(m))
```

```julia
using JuMP
using HiGHS

m = Model(HiGHS.Optimizer)

O = ["o1", "o2", "o3", "o4", "o5"]
W = ["w1", "w2", "w3", "w4", "w5", "w6", "w7"]
S = ["s"]
T = ["t"]

@variable(m, x[s in S, w in W] >= 0)  # Source to Owls
@variable(m, y[w in W, o in O] >= 0)  # Owls to Wands
@variable(m, z[o in O, t in T] >= 0)  # Wands to Sink

# Capacity constraints for owls to sink
@constraint(m, z["o1", "t"] <= 6)
@constraint(m, z["o2", "t"] <= 5)
@constraint(m, z["o3", "t"] <= 4)
@constraint(m, z["o4", "t"] <= 2)
@constraint(m, z["o5", "t"] <= 10)

# Ensure that each owl can carry at most one wand of each type
@constraint(m, [w in W, o in O], y[w, o] <= 1)

@constraint(m, [w in W], x["s", w] - sum(y[w, o] for o in O) == 0)
@constraint(m, [o in O], sum(y[w, o] for w in W) - z[o, "t"] == 0)

@objective(m, Max, sum(z[o, "t"] for o in O))  # Sum over the the sink

optimize!(m)

println("Maximum total number of wands delivered: ", objective_value(m))

owls = value.(z)
for i in 1:5
    println("Owl_$(i) carry total of $(owls["o$(i)", "t"]) wonds.")
end

wonds = value.(y)
result = zeros(7)
for i in 1:5
    for j in 1:7
        if wonds["w$(j)", "o$(i)"] === 1.0
            result[j] += 1
        end
    end
end

for i in 1:7
    println("$(result[i]) wond(s) of type $(i) shipped")
end
```
